```python
import pandas as pd

xk_df = pd.read_csv("../csv/xk.csv")
xk_df = xk_df[
    xk_df["Ngành"].isin(
        ["TỔNG TRỊ GIÁ", "Khu vực kinh tế trong nước", "Khu vực có vốn đầu tư NN"]
    )
]
xk_df
```

```python
import pandas as pd

nk_df = pd.read_csv("../csv/nk.csv")
nk_df = nk_df[
    nk_df["Ngành"].isin(
        ["TỔNG TRỊ GIÁ", "Khu vực kinh tế trong nước", "Khu vực có vốn đầu tư NN"]
    )
]
nk_df
```

```python
xnk_merge = pd.merge(
    left=xk_df,
    right=nk_df,
    on=["Ngành", "Date", "Month", "PreMonth"],
    suffixes=("-XK", "-NK"),
)
xnk_merge = xnk_merge.drop(columns=["Unnamed: 0-XK", "Unnamed: 0-NK"])
xnk_merge["Thặng dự tháng"] = xnk_merge["Total-M-XK"] - xnk_merge["Total-M-NK"]
xnk_merge["Thặng dự từ đầu năm"] = xnk_merge["Total-YOY-XK"] - xnk_merge["Total-YOY-NK"]
xnk_merge["Kim ngạch xnk tháng"] = xnk_merge["Total-M-XK"] + xnk_merge["Total-M-NK"]
xnk_merge["Kim ngạch xnk từ đầu năm"] = (
    xnk_merge["Total-YOY-XK"] + xnk_merge["Total-YOY-NK"]
)
xnk_merge.head(n=5)
```

```python
xnk_merge["Date"] = pd.to_datetime(xnk_merge["Date"], dayfirst=False)
xnk_merge["PreMonth"] = (
        xnk_merge["Date"]
        .apply(
            lambda x: (
                x.replace(month=x.month - 1, day=1)
                if x.month > 1
                else x.replace(year=x.year - 1, month=12, day=1)
            )
        )
        .dt.strftime("%Y-%m")
    )
xnk_merge["PreYearMonth"] = (
    xnk_merge["Date"]
    .apply(lambda x: x.replace(year=x.year - 1, day=1))
    .dt.strftime("%Y-%m")
)
xnk_merge.head(n=5)
```

```python
merge_df = pd.merge(
    left=xnk_merge,
    right=xnk_merge,
    left_on=["Ngành", "PreMonth"],
    right_on=["Ngành", "Month"],
    how="left",
)

merge_df["Thặng dự tháng - MOM"] = (
    100 * (merge_df["Thặng dự tháng_x"] / merge_df["Thặng dự tháng_y"]) - 100
)
# merge_df["Thặng dự từ đầu năm - MOM"] = (
#     100 * (merge_df["Thặng dự từ đầu năm_x"] / merge_df["Thặng dự từ đầu năm_y"]) - 100
# )
merge_df["Kim ngạch xnk tháng - MOM"] = (
    100 * (merge_df["Kim ngạch xnk tháng_x"] / merge_df["Kim ngạch xnk tháng_y"]) - 100
)
# merge_df["Kim ngạch xnk từ đầu năm - MOM"] = (
#     100
#     * (merge_df["Kim ngạch xnk từ đầu năm_x"] / merge_df["Kim ngạch xnk từ đầu năm_y"])
#     - 100
# )

merge_df = merge_df.drop(
    columns=[
        "Total-M-XK_y",
        "Total-YOY-XK_y",
        "YOY-XK_y",
        "YTD-YOY-XK_y",
        "Date_y",
        "Month_y",
        "PreMonth_y",
        "MOM-XK_y",
        "Total-M-NK_y",
        "Total-YOY-NK_y",
        "YOY-NK_y",
        "YTD-YOY-NK_y",
        "MOM-NK_y",
        "Thặng dự tháng_y",
        "Thặng dự từ đầu năm_y",
        "Kim ngạch xnk tháng_y",
        "Kim ngạch xnk từ đầu năm_y",
        "PreYearMonth_y",
    ]
)
merge_df = merge_df.rename(
    columns={
        "Total-M-XK_x": "Total-M-XK",
        "Total-YOY-XK_x": "Total-YOY-XK",
        "YOY-XK_x": "YOY-XK",
        "YTD-YOY-XK_x": "YTD-YOY-XK",
        "Date_x": "Date",
        "Month_x": "Month",
        "PreMonth_x": "PreMonth",
        "MOM-XK_x": "MOM-XK",
        "Total-M-NK_x": "Total-M-NK",
        "Total-YOY-NK_x": "Total-YOY-NK",
        "YOY-NK_x": "YOY-NK",
        "YTD-YOY-NK_x": "YTD-YOY-NK",
        "MOM-NK_x": "MOM-NK",
        "Thặng dự tháng_x": "Thặng dự tháng",
        "Thặng dự từ đầu năm_x": "Thặng dự từ đầu năm",
        "Kim ngạch xnk tháng_x": "Kim ngạch xnk tháng",
        "Kim ngạch xnk từ đầu năm_x": "Kim ngạch xnk từ đầu năm",
        "PreYearMonth_x": "PreYearMonth",
    }
)
xnk_merge = merge_df
merge_df[merge_df["Ngành"] == "TỔNG TRỊ GIÁ"].columns
```

```python
merge_df = pd.merge(
    left=xnk_merge,
    right=xnk_merge,
    left_on=["Ngành", "PreYearMonth"],
    right_on=["Ngành", "Month"],
    how="left",
)

merge_df["Thặng dự tháng - YOY"] = (
    100 * (merge_df["Thặng dự tháng_x"] / merge_df["Thặng dự tháng_y"]) - 100
)
merge_df["Thặng dự từ đầu năm - YOY"] = (
    100 * (merge_df["Thặng dự từ đầu năm_x"] / merge_df["Thặng dự từ đầu năm_y"]) - 100
)
merge_df["Kim ngạch xnk tháng - YOY"] = (
    100 * (merge_df["Kim ngạch xnk tháng_x"] / merge_df["Kim ngạch xnk tháng_y"]) - 100
)
merge_df["Kim ngạch xnk từ đầu năm - YOY"] = (
    100
    * (merge_df["Kim ngạch xnk từ đầu năm_x"] / merge_df["Kim ngạch xnk từ đầu năm_y"])
    - 100
)

merge_df = merge_df.drop(
    columns=[
        "Total-M-XK_y",
        "Total-YOY-XK_y",
        "YOY-XK_y",
        "YTD-YOY-XK_y",
        "Date_y",
        "Month_y",
        "PreMonth_y",
        "MOM-XK_y",
        "Total-M-NK_y",
        "Total-YOY-NK_y",
        "YOY-NK_y",
        "YTD-YOY-NK_y",
        "MOM-NK_y",
        "Thặng dự tháng_y",
        "Thặng dự từ đầu năm_y",
        "Kim ngạch xnk tháng_y",
        "Kim ngạch xnk từ đầu năm_y",
        "PreYearMonth_y",
        "Thặng dự tháng - MOM_y",
        "Kim ngạch xnk tháng - MOM_y",
    ]
)
merge_df = merge_df.rename(
    columns={
        "Total-M-XK_x": "Total-M-XK",
        "Total-YOY-XK_x": "Total-YOY-XK",
        "YOY-XK_x": "YOY-XK",
        "YTD-YOY-XK_x": "YTD-YOY-XK",
        "Date_x": "Date",
        "Month_x": "Month",
        "PreMonth_x": "PreMonth",
        "MOM-XK_x": "MOM-XK",
        "Total-M-NK_x": "Total-M-NK",
        "Total-YOY-NK_x": "Total-YOY-NK",
        "YOY-NK_x": "YOY-NK",
        "YTD-YOY-NK_x": "YTD-YOY-NK",
        "MOM-NK_x": "MOM-NK",
        "Thặng dự tháng_x": "Thặng dự tháng",
        "Thặng dự từ đầu năm_x": "Thặng dự từ đầu năm",
        "Kim ngạch xnk tháng_x": "Kim ngạch xnk tháng",
        "Kim ngạch xnk từ đầu năm_x": "Kim ngạch xnk từ đầu năm",
        "PreYearMonth_x": "PreYearMonth",
        "Thặng dự tháng - MOM_x": "Thặng dự tháng - MOM",
        "Kim ngạch xnk tháng - MOM_x": "Kim ngạch xnk tháng - MOM",
    }
)
xnk_merge = merge_df
merge_df[merge_df["Ngành"] == "TỔNG TRỊ GIÁ"].columns
```
