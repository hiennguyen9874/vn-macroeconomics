```python
def find_sheet_name(sheet_names):
    filtered_sheet_names = list(
        filter(
            lambda x: (
                "vdt" in x.lower()
                or "vđt" in x.lower()
                or "von" in x.lower()
                or "vnsnn" in x.lower()
            )
            and "nsnn" in x.lower()
            and ("quý" not in x.lower())
            and ("quy" not in x.lower())
            and ("txh" not in x.lower()),
            sheet_names,
        )
    )

    if len(filtered_sheet_names) == 1:
        return filtered_sheet_names[0]

    filtered_sheet_names = list(
        filter(
            lambda x: (
                "vdt" in x.lower()
                or "vđt" in x.lower()
                or "von" in x.lower()
                or "vnsnn" in x.lower()
            )
            and "thực hiện" in x.lower()
            and ("quý" not in x.lower())
            and ("quy" not in x.lower())
            and ("txh" not in x.lower()),
            sheet_names,
        )
    )

    if len(filtered_sheet_names) == 1:
        return filtered_sheet_names[0]

    filtered_sheet_names = list(
        filter(
            lambda x: (
                "vdt" in x.lower()
                or "vđt" in x.lower()
                or "von" in x.lower()
                or "vnsnn" in x.lower()
            )
            and ("quý" not in x.lower())
            and ("quy" not in x.lower())
            and ("txh" not in x.lower()),
            sheet_names,
        )
    )

    if len(filtered_sheet_names) == 1:
        return filtered_sheet_names[0]

    filtered_sheet_names = list(
        filter(
            lambda x: (
                "vdt" in x.lower()
                or "vđt" in x.lower()
                or "von" in x.lower()
                or "vnsnn" in x.lower()
                or "dtu" in x.lower()
            )
            and ("quý" not in x.lower())
            and ("quy" not in x.lower())
            and ("txh" not in x.lower()),
            sheet_names,
        )
    )

    if len(filtered_sheet_names) == 1:
        return filtered_sheet_names[0]

    if len(sheet_names) == 43:
        return "20"
    
    if len(sheet_names) == 13:
        return "Sheet1"

    raise RuntimeError(sheet_names, filtered_sheet_names)
```

```python
def get_vdt(xl, sheet_name, date):
    df = xl.parse(sheet_name)

    df = df.dropna(axis=1, how="all")

    df[df.columns[0]] = (
        df[df.columns[0]]
        .replace(r"\s+", " ", regex=True)
        .replace(r"\n", " ", regex=True)
        .str.strip()
    )
    df[df.columns[1]] = (
        df[df.columns[1]]
        .replace(r"\s+", " ", regex=True)
        .replace(r"\n", " ", regex=True)
        .str.strip()
    )

    if (
        len(
            df[df.columns[1]][
                (df[df.columns[1]] == "Trung ương") | (df[df.columns[1]] == "Central")
            ].index
        )
        > 0
    ):
        df[df.columns[1]] = df[df.columns[1]].fillna(df[df.columns[0]])
        df = df.drop(columns=[df.columns[0]])
    df = df.reset_index(drop=True)

    df = df.drop(
        list(
            range(
                df[df.columns[0]][
                    (df[df.columns[0]] == "TỔNG SỐ") | ((df[df.columns[0]] == "TOTAL"))
                ].index[0]
            )
        )
    )
    df = df.reset_index(drop=True)

    if len(df[df.columns[0]][df[df.columns[0]] == "TOTAL"].index) > 0:
        df[df.columns[0]] = df[df.columns[0]].map(en_to_vn)

    if len(df.columns) == 6:
        df = df.drop(columns=[df.columns[1]])
        df.columns = ["Phân Loại", "Total-M", "Total-YOY", "YOP", "YTD-YOY"]

        df = df[df["Total-M"].notna()].replace("-", None)
        df = df[df["Total-YOY"].notna()].replace("-", None)
        df = df[df["YOP"].notna()].replace("-", None)
        df = df[df["YTD-YOY"].notna()].replace("-", None)

        df["YTD-YOY"] = df["YTD-YOY"] - 100
    elif len(df.columns) == 5 and int(date.split("-")[1]) == 1:
        df = df.drop(columns=[df.columns[1]])
        df.columns = ["Phân Loại", "Total-M", "YOP", "YTD-YOY"]
        df["Total-YOY"] = df["Total-M"]

        df = df[df["Total-M"].notna()].replace("-", None)
        df = df[df["Total-YOY"].notna()].replace("-", None)
        df = df[df["YOP"].notna()].replace("-", None)
        df = df[df["YTD-YOY"].notna()].replace("-", None)

        df["YTD-YOY"] = df["YTD-YOY"] - 100

    else:
        raise RuntimeError

    df["Date"] = date
    df["Date"] = pd.to_datetime(df["Date"], dayfirst=False)

    return df


xl = pd.ExcelFile("../datas/2015-05-29-Bieu-5-2015-1.xlsx")

get_vdt(xl, find_sheet_name(xl.sheet_names), "2015-05-29")
```

```python
import os
import contextlib

dfs_vdt_origin = []
for file_name in sorted(
    os.listdir("../datas"),
    key=lambda filename: f"{filename.split('-')[0]}-{filename.split('-')[1]}-{filename.split('-')[2]}",
    reverse=True,
):
    if file_name.startswith(".~"):
        continue

    try:
        file_path = os.path.join("../datas", file_name)

        year = int(file_name.split("-")[0])
        month = int(file_name.split("-")[1])
        day = int(file_name.split("-")[2])

        if year < 2012:
            continue

        xl = pd.ExcelFile(file_path)

        df_xk = get_vdt(xl, find_sheet_name(xl.sheet_names), f"{year}-{month}-{day}")

        dfs_vdt_origin.append(df_xk)
    except Exception as e:
        print(file_name)
        raise e
```

```python
dfs_vdt = pd.concat(dfs_vdt_origin, ignore_index=True, axis=0)
dfs_vdt["Date"] = pd.to_datetime(dfs_vdt["Date"], dayfirst=False)
dfs_vdt
```
